```python
import matplotlib.pyplot as plt
import numpy as np

squares = [1,4,9,16,25]
plt.plot(squares, linewidth=5)
plt.title('Numeros cuadrados', fontsize=18)
plt.xlabel('Valor', fontsize=11)
plt.ylabel('Valor cuadrado', fontsize=11)
plt.tick_params(axis='ambos', labelsize=12)
plt.show()
```

```python
import matplotlib.pyplot as plt2
import numpy as np

#definimos nuestra funcion
def f(x):
    return np.exp(-x ** 2)
#creamos un vector con los puntos que le pasaremos a la funcion previamente creada
x = np.linspace(-1, 5, num=30)
#representaremos la funcion utilizando el objeto plt de matplotlib
plt2.xlabel('Eje $x$', fontsize=11)
plt2.ylabel('$f(x)$', fontsize=11)
plt2.title('Funcion $f(x)$')
plt2.grid(True)
plt2.legend('Funcion')
plt2.plot(x, f(x), label='Funcion f(x)')
plt2.show()
```

```python
import matplotlib.pyplot as plt3
import numpy as np3

#grafico de puntos con matplotlib
N = 100
x1 = np3.random.randn(N) # creando el vector x
y1 = np3.random.randn(N) # creando el vector y

s = 50 + 50 * np3.random.randn(N) #variable para modificar el tamao(size)
c = np3.random.randn(N) #variable para modificar el color(color)

plt3.scatter(x1, y1, s=s, c=c, cmap=plt3.cm.Blues)
plt3.grid(True)
plt3.colorbar()
plt3.scatter(x1, y1)
plt3.show()
```

```python
x = np.linspace(0, 5, 10) #conjunto de puntos
y = x ** 2 #funcion 
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #eje principal
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) #eje secundario

#figura principal
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Ej OOP')
#insertada
axes2.plot(y, x, 'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('insertado')
plt.show()
```

```python
##ejemplo con mas de una figura
fig, axes = plt.subplots(nrows=1, ncols=2)
for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('titulo')

fig.tight_layout()
plt.show()
```
